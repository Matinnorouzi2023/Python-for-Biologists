```python
import matplotlib.pyplot as plt
```

```python
r = ['A','T','G','C']
c = [30, 20, 10, 40]
```

```python
plt.bar(r, c)
plt.show()
```

```python
plt.pie(c, labels=r)
plt.show()
```

```python
plt.pie(c, labels=r, colors=['gold','red','green','blue'], explode=(0,0.1,0,0), autopct='%0.1f%%')
plt.legend(loc='best')
plt.show()
```

```python
import pandas as pd
```

```python
df = pd.read_csv('d:AAC.csv')
df
```

```python
df.shape
```

```python
df.head(3)
```

```python
sum(df.n)
```

```python
m = df.n / 110
p = round(m * 100, 1)
```

```python
df['Percent'] = p
```

```python
df.head()
```

```python
plt.figure(figsize=(20,10))
plt.bar(df.AminoAcid, df.n)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.bar(df.AminoAcid, df.Percent)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.pie(df.n, labels=df.AminoAcid)
plt.show()
```

```python
plt.boxplot(df.n)
plt.show()
```

```python
df.n
```

```python
plt.figure(figsize=(20,10))

plt.scatter(df.AminoAcid, df.Percent)
plt.xlabel('AminoAcid')
plt.ylabel('Percent')
plt.title('AminoAcid Composition', fontsize=20)
plt.show()
```
