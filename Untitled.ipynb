```python
# Evolutionary relationship visualization
from Bio import Phylo
import matplotlib.pyplot as plt
from io import StringIO
import os

def plot_phylogeny_tree(file_path="phylogeny.nwk"):
    """Plot phylogenetic tree from Newick file or create sample"""
    
    try:
        # سعی در خواندن فایل
        tree = Phylo.read(file_path, "newick")
        print(f"Successfully loaded tree from {file_path}")
        
    except FileNotFoundError:
        print(f"File {file_path} not found. Creating sample tree...")
        
        # ایجاد درخت نمونه
        sample_newick = "(((Human:0.1,Chimpanzee:0.08):0.05,Gorilla:0.12):0.03,((Orangutan:0.15,(Macaque:0.2,Baboon:0.18):0.1):0.05,Lemur:0.25));"
        
        tree = Phylo.read(StringIO(sample_newick), "newick")
        
        # ذخیره نمونه برای استفاده بعدی
        Phylo.write(tree, file_path, "newick")
        print(f"Sample tree saved to {file_path}")
    
    # رسم درخت
    plt.figure(figsize=(14, 10))
    Phylo.draw(tree,
               label_func=lambda x: x.name if x.name else "",
               branch_labels=lambda c: f"{c.branch_length:.2f}" if c.branch_length else "",
               do_show=False)
    
    plt.title("Phylogenetic Tree - Evolutionary Relationships", fontsize=16, pad=20)
    plt.tight_layout()
    plt.show()
    
    return tree

# اجرای تابع
tree = plot_phylogeny_tree()

# نمایش اطلاعات اضافی درباره درخت
print(f"\nTree information:")
print(f"Number of terminal nodes (leaves): {tree.count_terminals()}")
print(f"Total number of nodes: {tree.count_clades()}")
print(f"Tree depth: {tree.depth()}")
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

def load_or_create_expression_data(file_path="gene_expression.csv"):
    """بارگیری فایل یا ایجاد داده‌های نمونه"""
    
    if os.path.exists(file_path):
        print(f"Loading data from {file_path}")
        return pd.read_csv(file_path, index_col=0)
    else:
        print(f"File {file_path} not found. Creating sample data...")
        return create_sample_data_and_save(file_path)

def create_sample_data_and_save(file_path):
    """ایجاد داده‌های نمونه و ذخیره فایل"""
    np.random.seed(42)
    
    # ایجاد داده‌های واقعی‌تر
    n_genes = 30
    n_samples = 12
    
    genes = [f'Gene_{i}' for i in range(1, n_genes + 1)]
    samples = [f'Control_{i}' for i in range(1, 7)] + [f'Treatment_{i}' for i in range(1, 7)]
    
    # ایجاد داده با الگوهای بیولوژیکی
    data = np.random.gamma(shape=2, scale=1, size=(n_genes, n_samples))
    
    # اضافه کردن تفاوت بین کنترل و تریتمنت
    treatment_indices = [i for i, sample in enumerate(samples) if 'Treatment' in sample]
    data[5:15, treatment_indices] *= 3  # افزایش بیان در تریتمنت
    data[20:25, treatment_indices] *= 0.3  # کاهش بیان در تریتمنت
    
    df = pd.DataFrame(data, index=genes, columns=samples)
    df.to_csv(file_path)
    print(f"Sample data saved to {file_path}")
    
    return df

# بارگیری یا ایجاد داده‌ها
expression_data = load_or_create_expression_data()

print(f"Data shape: {expression_data.shape}")
print("\nFirst few rows:")
print(expression_data.head())

# رسم heatmap
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

# Heatmap ساده
sns.heatmap(expression_data, cmap="viridis", ax=ax1,
            cbar_kws={'label': 'Expression Level'})
ax1.set_title("Gene Expression Heatmap", fontsize=14)
ax1.set_xlabel("Samples")
ax1.set_ylabel("Genes")

# Heatmap با آنوتیشن
sns.heatmap(expression_data, cmap="coolwarm", ax=ax2,
            annot=False,  # برای خوانایی آنوتیشن را غیرفعال کردیم
            center=expression_data.values.mean(),
            cbar_kws={'label': 'Expression Level'})
ax2.set_title("Gene Expression Heatmap (Centered)", fontsize=14)
ax2.set_xlabel("Samples")
ax2.set_ylabel("Genes")

plt.tight_layout()
plt.show()

# نمایش آمار توصیفی
print("\nDescriptive Statistics:")
print(expression_data.describe())
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import umap.umap_ as umap
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

print("Single-cell RNA-seq Analysis with Alternative Approach")

def create_sample_single_cell_data():
    """ایجاد داده‌های نمونه single-cell"""
    np.random.seed(42)
    
    # پارامترها
    n_cells = 500
    n_genes = 2000
    
    # ایجاد انواع سلول
    cell_types = ['T_cells', 'B_cells', 'Monocytes', 'NK_cells', 'Dendritic']
    n_per_type = n_cells // len(cell_types)
    
    # ایجاد ماتریس بیان پایه
    data = np.random.negative_binomial(5, 0.3, (n_cells, n_genes))
    
    # اضافه کردن signatureهای خاص برای هر نوع سلول
    marker_regions = {
        'T_cells': (100, 200),
        'B_cells': (200, 300),
        'Monocytes': (300, 400),
        'NK_cells': (400, 500),
        'Dendritic': (500, 600)
    }
    
    cell_labels = []
    batch_labels = []
    
    for i, cell_type in enumerate(cell_types):
        start_idx = i * n_per_type
        end_idx = (i + 1) * n_per_type
        cell_labels.extend([cell_type] * n_per_type)
        batch_labels.extend(['Batch1'] * (n_per_type//2) + ['Batch2'] * (n_per_type//2))
        
        # افزایش بیان ژن‌های مارکر
        start_gene, end_gene = marker_regions[cell_type]
        data[start_idx:end_idx, start_gene:end_gene] += np.random.poisson(10, size=(n_per_type, 100))
    
    # ایجاد نام‌ها
    cell_ids = [f'Cell_{i:04d}' for i in range(n_cells)]
    gene_ids = [f'Gene_{i:05d}' for i in range(n_genes)]
    
    # ایجاد DataFrame
    adata_df = pd.DataFrame(data, index=cell_ids, columns=gene_ids)
    metadata_df = pd.DataFrame({
        'cell_type': cell_labels,
        'batch': batch_labels
    }, index=cell_ids)
    
    return adata_df, metadata_df

def preprocess_sc_data(expression_df, n_top_genes=1000):
    """پیش‌پردازش داده‌های single-cell"""
    print("Preprocessing data...")
    
    # 1. Normalization (CPM)
    cell_totals = expression_df.sum(axis=1)
    normalized = expression_df.div(cell_totals, axis=0) * 10000
    
    # 2. Log transformation
    log_transformed = np.log1p(normalized)
    
    # 3. Highly variable genes selection
    gene_means = log_transformed.mean(axis=0)
    gene_vars = log_transformed.var(axis=0)
    dispersion = gene_vars / gene_means
    
    # انتخاب ژن‌های highly variable
    hvg_indices = dispersion.nlargest(n_top_genes).index
    hvg_data = log_transformed[hvg_indices]
    
    print(f"Selected {len(hvg_indices)} highly variable genes")
    return hvg_data, hvg_indices

def perform_dimension_reduction(data, metadata):
    """انجام کاهش ابعاد با روش‌های مختلف"""
    print("Performing dimension reduction...")
    
    # استانداردسازی
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(data)
    
    # PCA
    pca = PCA(n_components=50, random_state=42)
    pca_result = pca.fit_transform(scaled_data)
    
    # UMAP
    umap_reducer = umap.UMAP(n_components=2, random_state=42, n_neighbors=15, min_dist=0.1)
    umap_result = umap_reducer.fit_transform(scaled_data)
    
    # t-SNE
    tsne = TSNE(n_components=2, random_state=42, perplexity=30, learning_rate=200)
    tsne_result = tsne.fit_transform(pca_result[:, :30])
    
    # ایجاد DataFrame نتایج
    results_df = pd.DataFrame({
        'UMAP1': umap_result[:, 0],
        'UMAP2': umap_result[:, 1],
        'tSNE1': tsne_result[:, 0],
        'tSNE2': tsne_result[:, 1],
        'PC1': pca_result[:, 0],
        'PC2': pca_result[:, 1],
        'cell_type': metadata['cell_type'],
        'batch': metadata['batch']
    })
    
    return results_df, pca

# ایجاد داده‌های نمونه
print("Creating sample single-cell data...")
expression_data, metadata = create_sample_single_cell_data()
print(f"Created dataset: {expression_data.shape[0]} cells × {expression_data.shape[1]} genes")

# پیش‌پردازش
processed_data, hvg_genes = preprocess_sc_data(expression_data)

# کاهش ابعاد
results_df, pca_model = perform_dimension_reduction(processed_data, metadata)

# رسم نتایج
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# پالت رنگ
cell_type_palette = {'T_cells': '#1f77b4', 'B_cells': '#ff7f0e', 
                     'Monocytes': '#2ca02c', 'NK_cells': '#d62728', 
                     'Dendritic': '#9467bd'}
batch_palette = {'Batch1': '#8c564b', 'Batch2': '#e377c2'}

# UMAP - Cell Types
for cell_type in results_df['cell_type'].unique():
    subset = results_df[results_df['cell_type'] == cell_type]
    axes[0,0].scatter(subset['UMAP1'], subset['UMAP2'], 
                     label=cell_type, color=cell_type_palette[cell_type],
                     s=30, alpha=0.7, edgecolors='white', linewidth=0.5)
axes[0,0].set_title('UMAP - Cell Types', fontsize=14, fontweight='bold')
axes[0,0].set_xlabel('UMAP1')
axes[0,0].set_ylabel('UMAP2')
axes[0,0].legend()

# UMAP - Batch
for batch in results_df['batch'].unique():
    subset = results_df[results_df['batch'] == batch]
    axes[0,1].scatter(subset['UMAP1'], subset['UMAP2'], 
                     label=batch, color=batch_palette[batch],
                     s=30, alpha=0.7, edgecolors='white', linewidth=0.5)
axes[0,1].set_title('UMAP - Batch Effect', fontsize=14, fontweight='bold')
axes[0,1].set_xlabel('UMAP1')
axes[0,1].set_ylabel('UMAP2')
axes[0,1].legend()

# t-SNE - Cell Types
for cell_type in results_df['cell_type'].unique():
    subset = results_df[results_df['cell_type'] == cell_type]
    axes[0,2].scatter(subset['tSNE1'], subset['tSNE2'], 
                     label=cell_type, color=cell_type_palette[cell_type],
                     s=30, alpha=0.7, edgecolors='white', linewidth=0.5)
axes[0,2].set_title('t-SNE - Cell Types', fontsize=14, fontweight='bold')
axes[0,2].set_xlabel('t-SNE1')
axes[0,2].set_ylabel('t-SNE2')
axes[0,2].legend()

# PCA - Cell Types
for cell_type in results_df['cell_type'].unique():
    subset = results_df[results_df['cell_type'] == cell_type]
    axes[1,0].scatter(subset['PC1'], subset['PC2'], 
                     label=cell_type, color=cell_type_palette[cell_type],
                     s=30, alpha=0.7, edgecolors='white', linewidth=0.5)
axes[1,0].set_title('PCA - Cell Types (PC1 vs PC2)', fontsize=14, fontweight='bold')
axes[1,0].set_xlabel('Principal Component 1')
axes[1,0].set_ylabel('Principal Component 2')
axes[1,0].legend()

# Variance Explained
components = range(1, 21)
explained_variance = pca_model.explained_variance_ratio_[:20]
axes[1,1].plot(components, explained_variance, 'o-', color='#2ca02c', linewidth=2, markersize=6)
axes[1,1].set_title('PCA Explained Variance', fontsize=14, fontweight='bold')
axes[1,1].set_xlabel('Principal Component')
axes[1,1].set_ylabel('Explained Variance Ratio')
axes[1,1].grid(True, alpha=0.3)

# Cumulative Variance
cumulative_variance = np.cumsum(pca_model.explained_variance_ratio_[:20])
axes[1,2].plot(components, cumulative_variance, 's-', color='#d62728', linewidth=2, markersize=6)
axes[1,2].set_title('Cumulative Explained Variance', fontsize=14, fontweight='bold')
axes[1,2].set_xlabel('Principal Component')
axes[1,2].set_ylabel('Cumulative Variance Ratio')
axes[1,2].grid(True, alpha=0.3)
axes[1,2].axhline(y=0.8, color='gray', linestyle='--', alpha=0.7)
axes[1,2].text(10, 0.82, '80% variance', fontsize=10)

plt.tight_layout()
plt.show()

# نمایش آمار
print(f"\n📊 Dataset Statistics:")
print(f"• Total cells: {expression_data.shape[0]}")
print(f"• Total genes: {expression_data.shape[1]}")
print(f"• Highly variable genes: {len(hvg_genes)}")
print(f"• Cell types: {', '.join(metadata['cell_type'].unique())}")
print(f"• First 5 PCs explain: {pca_model.explained_variance_ratio_[:5].sum():.2%} of variance")
print(f"• First 10 PCs explain: {pca_model.explained_variance_ratio_[:10].sum():.2%} of variance")

# Heatmap بیان ژن‌های مارکر
print("\nCreating marker gene heatmap...")
marker_genes_to_plot = ['Gene_00100', 'Gene_00150', 'Gene_00200', 'Gene_00250', 'Gene_00300']

# بررسی وجود ژن‌ها
available_markers = [gene for gene in marker_genes_to_plot if gene in expression_data.columns]

if available_markers:
    # انتخاب 50 سلول برای heatmap
    heatmap_data = expression_data[available_markers].iloc[:50]
    
    # نرمال‌سازی برای heatmap
    heatmap_data_normalized = (heatmap_data - heatmap_data.mean()) / heatmap_data.std()
    
    plt.figure(figsize=(12, 8))
    sns.heatmap(heatmap_data_normalized.T, 
                cmap='coolwarm', 
                center=0,
                yticklabels=available_markers,
                xticklabels=False,
                cbar_kws={'label': 'Z-score'})
    plt.title('Marker Gene Expression Heatmap', fontsize=16, fontweight='bold')
    plt.xlabel('Cells')
    plt.ylabel('Genes')
    plt.tight_layout()
    plt.show()
else:
    print("Marker genes not found in dataset")

print("✅ Single-cell RNA-seq analysis completed successfully!")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import warnings
warnings.filterwarnings('ignore')

print("Genome Browser Visualization - Alternative Approach")

def create_sample_genome_data():
    """ایجاد داده‌های نمونه ژنومی"""
    # ایجاد داده‌های coverage شبیه bigwig
    genome_length = 10000
    positions = np.arange(genome_length)
    
    # ایجاد الگوهای coverage مختلف
    coverage = np.zeros(genome_length)
    
    # اضافه کردن peakها در مناطق مختلف
    peaks = [
        (1000, 1500, 5.0),   # (start, end, height)
        (3000, 3500, 8.0),
        (6000, 7000, 6.5),
        (8000, 8500, 7.2)
    ]
    
    for start, end, height in peaks:
        peak_length = end - start
        peak_data = np.random.normal(height, 0.5, peak_length)
        coverage[start:end] = np.maximum(coverage[start:end], peak_data)
    
    # اضافه کردن noise پایه
    coverage += np.random.exponential(0.1, genome_length)
    
    # ایجاد داده‌های ژن
    genes = [
        {"name": "Gene_A", "start": 1200, "end": 1800, "strand": "+", "color": "red"},
        {"name": "Gene_B", "start": 3200, "end": 3800, "strand": "-", "color": "blue"},
        {"name": "Gene_C", "start": 6200, "end": 7200, "strand": "+", "color": "green"},
        {"name": "Gene_D", "start": 8200, "end": 8800, "strand": "-", "color": "purple"}
    ]
    
    return positions, coverage, genes

def plot_genome_browser(positions, coverage, genes, region_start=0, region_end=10000):
    """رسم genome browser با matplotlib"""
    
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 10), 
                                       gridspec_kw={'height_ratios': [3, 1, 1]})
    
    # 1. Track Coverage (شبیه bigwig)
    ax1.fill_between(positions, coverage, alpha=0.7, color='blue', label='Coverage')
    ax1.plot(positions, coverage, 'b-', alpha=0.8, linewidth=1)
    ax1.set_ylabel('Coverage\nDepth', fontsize=12)
    ax1.set_title('Genome Browser View', fontsize=16, fontweight='bold')
    ax1.grid(True, alpha=0.3)
    ax1.legend()
    
    # اضافه کردن annotation برای peakها
    for i, pos in enumerate(positions[::1000]):
        if coverage[pos] > 2:
            ax1.annotate(f'Peak', xy=(pos, coverage[pos]), 
                        xytext=(10, 10), textcoords='offset points',
                        bbox=dict(boxstyle='round,pad=0.3', fc='yellow', alpha=0.7),
                        arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0'))
    
    # 2. Track ژن‌ها
    ax2.set_ylabel('Genes', fontsize=12)
    ax2.set_ylim(0, 1)
    
    for i, gene in enumerate(genes):
        # محاسبه موقعیت عمودی
        y_pos = 0.5
        
        # رسم جعبه ژن
        gene_width = gene["end"] - gene["start"]
        rect = Rectangle((gene["start"], y_pos - 0.2), gene_width, 0.4,
                       facecolor=gene["color"], alpha=0.7, edgecolor='black')
        ax2.add_patch(rect)
        
        # اضافه کردن نام ژن
        ax2.text(gene["start"] + gene_width/2, y_pos + 0.3, gene["name"], 
                ha='center', va='bottom', fontsize=10, fontweight='bold')
        
        # رسم strand (جهت)
        arrow_length = min(100, gene_width * 0.2)
        if gene["strand"] == "+":
            ax2.arrow(gene["start"] + gene_width, y_pos, arrow_length, 0, 
                     head_width=0.1, head_length=50, fc='black', ec='black')
        else:
            ax2.arrow(gene["start"], y_pos, -arrow_length, 0, 
                     head_width=0.1, head_length=50, fc='black', ec='black')
    
    ax2.set_xlim(region_start, region_end)
    ax2.set_xticks([])
    
    # 3. Track مقیاس (scale bar)
    ax3.set_ylabel('Scale', fontsize=12)
    ax3.set_ylim(0, 1)
    
    # رسم scale bar
    scale_length = 1000
    scale_start = (region_start + region_end) / 2 - scale_length / 2
    ax3.hlines(0.5, scale_start, scale_start + scale_length, colors='black', linewidth=3)
    ax3.vlines([scale_start, scale_start + scale_length], 0.4, 0.6, colors='black', linewidth=2)
    ax3.text(scale_start + scale_length / 2, 0.7, '1 kb', 
            ha='center', va='bottom', fontsize=12, fontweight='bold')
    
    ax3.set_xlim(region_start, region_end)
    ax3.set_xlabel('Genomic Position (bp)', fontsize=12)
    ax3.set_xticks(np.arange(region_start, region_end + 1, 2000))
    ax3.set_xticklabels([f'{int(x/1000)}k' for x in np.arange(region_start, region_end + 1, 2000)])
    
    plt.tight_layout()
    plt.show()
    
    return fig

# ایجاد داده‌های نمونه
print("Creating sample genome data...")
positions, coverage, genes = create_sample_genome_data()

print(f"Generated data: {len(positions)} positions, {len(genes)} genes")
print("Genes in dataset:")
for gene in genes:
    print(f"  - {gene['name']}: {gene['start']}-{gene['end']} ({gene['strand']})")

# رسم genome browser
fig = plot_genome_browser(positions, coverage, genes)

# نمایش آمار داده‌ها
print(f"\n📊 Coverage Statistics:")
print(f"• Mean coverage: {coverage.mean():.2f}")
print(f"• Max coverage: {coverage.max():.2f}")
print(f"• Regions with coverage > 2: {(coverage > 2).sum()} bp")
print(f"• Genome size: {len(positions)} bp")
```
